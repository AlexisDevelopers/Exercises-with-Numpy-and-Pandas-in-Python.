## 1.3 
De la siguiente matriz cálcula:  
*  El máximo de **cada fila**, el resultado debe ser un array de 5 elementos.
*  La media de **cada columna**, el resultado debe ser un array de 6 elementos.
*  La desviación típica **total** (de todos los elementos de la matriz), el resultado es un escalar.

<center>
<img src="https://drive.google.com/uc?export=view&id=1orExRlmRDPkNhj8uo9tiORYg7B8992NC">

# Ejercicios Pandas

```python
import pandas as pd
```

```python
# Air Passengers: Dataset de Box & Jenkins con el número total de pasajeros internacionales mensuales de 1949 a 1960
air_passengers_share = "https://drive.google.com/file/d/1UQuel7m_98HEzKkDiWlxXDcr-ftimvqT/view?usp=sharing"
url ='https://drive.google.com/uc?id=' + air_passengers_share.split('/')[-2]
air_df = pd.read_csv(url).rename(columns={"Thousands of Passengers":"Pasajeros"})
air_df.head()
```

## 2.1 
Establece un DatetimeIndex para las filas del DataFrame dado usando la columna `Month`. Después **elimina la columna** de `'Month'` y establece la `freq` a **Mensual Start**

**Pista**: Utiliza `pd.to_datetime` para convertir de str a Timestamps y luego establece esa columna de índice.

```python
air_df['Month'] = pd.to_datetime(air_df['Month'])
air_df.set_index('Month', drop=True, inplace=True)
air_df.index.freq = 'MS'
```

## 2.2
Calcula y grafica mediante el uso de **ventanas móviles** de 3 y 6 meses (es decir, dos gráficas distintas, una para ventana de 3 meses y la otra para las ventanas de 6 meses):  
  * Media móvil de pasajeros en la ventana
  * Máximo móvil de pasajeros en la ventana
  * Mínimo móvil de pasajeros en la ventana
  * Rango (máximo - mínimo) móvil móvil de pasajeros en la ventana

Las gráficas deberán mostrar las 4 curvas "móviles" (media, máximo, mínimo y rango) junto con la de pasajeros mensuales, es decir, **5 curvas en total**

```python
df_23 = air_df.copy()
display(df_23[['Pasajeros']].rolling(window=7).mean().head(15))

df_23['Media(Month_3)'] = df_23['Pasajeros'].rolling(window=3).mean()
df_23['Maximo(Month_3)'] = df_23['Pasajeros'].rolling(window=3).max()
df_23['Minimo(Month_3)'] = df_23['Pasajeros'].rolling(window=3).min()
df_23['Rango(Month_3)'] = df_23['Maximo(Month_3)'] - df_23['Minimo(Month_3)']
df_23[[ 'Pasajeros', 'Media(Month_3)', 'Maximo(Month_3)', 'Minimo(Month_3)', 'Rango(Month_3)']].plot(figsize=(12,5)).autoscale(axis='x',tight=True)

df_23['Media(Month_6)'] = df_23['Pasajeros'].rolling(window=3).mean()
df_23['Maximo(Month_6)'] = df_23['Pasajeros'].rolling(window=6).max()
df_23['Minimo(Month_6)'] = df_23['Pasajeros'].rolling(window=6).min()
df_23['Rango(Month_6)'] = df_23['Maximo(Month_6)'] - df_23['Minimo(Month_6)']
df_23[[ 'Pasajeros', 'Media(Month_6)', 'Maximo(Month_6)', 'Minimo(Month_6)', 'Rango(Month_6)']].plot(figsize=(12,5)).autoscale(axis='x',tight=True)

```

## 2.3 
Calcula y grafica (una sola gráfica) los siguientes valores **trimestrales** (es decir Q1, Q2, Q3, Q4 stándar, 4 valores por cada año):  
  * El valor medio de pasajeros
  * El envoltorio de +1/-1 desviación típica del numero de viajeros  
    * Esto significa calular 2 curvas trimestrales:
      * Una con valor `media + desviación tipica` 
      * Otra con valor  `media - desviación tpica`

En la gráfica deberán aparecer **3 curvas**, la curva de los valores medios trimestrales, la curva de la media más la desviación típica trimestral y la curva de la media menos la desviación típica trimestral
   
**Notas**  
  * Copia el DataFrame dejando el original air_df sin modificar  
    * Por ejemplo haced `df_24 = air_df.copy()` al inicio  
  * Este ejercicio **no son medias moviles**

```python
df_24 = air_df.copy()

# Visualizacion de los valores Trimestrales
display(df_24['Pasajeros'].resample(rule='Q').mean().plot(figsize=(16,6), legend=True, label='Valores trimestrales', title='El valor medio de pasajeros'))
display(((df_24['Pasajeros'].resample(rule='Q').mean()) + (df_24['Pasajeros'].resample(rule='Q').std())).plot(figsize=(16,6), legend=True, label='Media más la desviación típica trimestral'))
display(((df_24['Pasajeros'].resample(rule='Q').mean()) - (df_24['Pasajeros'].resample(rule='Q').std())).plot(figsize=(16,6), legend=True, label='Media menos la desviación típica trimestral'))
```
